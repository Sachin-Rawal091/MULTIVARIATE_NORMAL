```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
weight = np.random.normal(loc=70,scale=5,size=500)
height = (-2*weight + 5)+np.random.normal(loc=20,scale=9,size=500)
data = pd.DataFrame({"Weight (Kgs)":weight,"Height (Cms)":height})
data
```

```python
plt.xlabel("Weight (Kgs)")
plt.ylabel("Height (Cms)")
plt.scatter(data['Weight (Kgs)'],data['Height (Cms)'])
```

```python
print(np.mean(data['Height (Cms)']))
print(np.mean(data['Weight (Kgs)']))
```
